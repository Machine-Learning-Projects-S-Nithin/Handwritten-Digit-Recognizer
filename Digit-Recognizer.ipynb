```python
train_data=pd.read_csv('Data/train.csv')
test_data=pd.read_csv('Data/test.csv')
```

```python
train_x=train_data.iloc[:,1:].values
train_y=train_data.iloc[:,0].values
```

```python
test_x=test_data.iloc[:,:].values
```

```python
train_x=np.divide(train_x,255)
test_x=np.divide(test_x,255)
```

```python
def test_model(model,trials):
    total_score=0
    for trial in range(trials):
        x_train, x_test, y_train, y_test=train_test_split(train_x,train_y, test_size=0.2, random_state=42)
        model.fit(x_train, y_train)
        total_score+=r2_score(y_test,model.predict(x_test))
    print("Average accuracy of the model: ",np.round(total_score/trials,decimals=4))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr_model_1=LogisticRegression(max_iter=10000)

test_model(lr_model_1, 1)
```

   # Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier

RFC_model=RandomForestClassifier()

test_model(RFC_model, 1)
```

   # Decision Tree classifier

```python
from sklearn.tree import DecisionTreeClassifier

DTC_model=DecisionTreeClassifier()

test_model(DTC_model,10)
```

# Naive Bayes classifier

```python
from sklearn.naive_bayes import GaussianNB

NBC_model=GaussianNB()

test_model(NBC_model,10)
```

# K Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier

KNN_model=KNeighborsClassifier()

test_model(KNN_model, 1)
```

# Convolutional Neural Network

```python
X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size = 0.2, random_state=42)
X_train = X_train.reshape(-1, 28, 28)
X_test = X_test.reshape(-1, 28, 28)
X = test_x.reshape(-1, 28, 28)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from keras.layers.advanced_activations import LeakyReLU

cnn_model=Sequential()

#Layer 1
cnn_model.add(Conv2D(128, kernel_size=(3,3), input_shape=(28,28,1), activation='relu', padding='same'))
cnn_model.add(LeakyReLU(alpha=0.1))
cnn_model.add(MaxPooling2D((2,2), padding='same'))

#Layer 2
cnn_model.add(Conv2D(128, kernel_size=(3,3), activation='relu', padding='same'))
cnn_model.add(LeakyReLU(alpha=0.1))
cnn_model.add(MaxPooling2D((2,2), padding='same'))
          
#Layer 3         
cnn_model.add(Flatten())
cnn_model.add(Dense(128, activation='linear'))

#Output layer
cnn_model.add(LeakyReLU(alpha=0.1))
cnn_model.add(Dense(10, activation='softmax'))

cnn_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
hist=cnn_model.fit(X_train.reshape(-1,28,28,1), y_train, batch_size = 32, epochs = 10, validation_data=(X_test.reshape(-1,28,28,1), y_test))


val_loss, val_acc = cnn_model.evaluate(X_test.reshape(-1, 28, 28, 1), y_test)

print("Loss: ",val_loss)
print("Accuracy : ",val_acc)
```
