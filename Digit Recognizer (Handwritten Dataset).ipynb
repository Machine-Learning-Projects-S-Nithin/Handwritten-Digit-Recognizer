```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
from tensorflow.keras.models import load_model

cnn_model=load_model('cnn_model')
```

```python
def print_image(img):
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
def print_data():
    for number in os.listdir('Handwritten Numbers'):
        path=os.path.join('Handwritten Numbers', number)
        image=cv.imread(path, cv.IMREAD_GRAYSCALE)
        plt.imshow(image)
        plt.show()
```

```python
print_data()
```

```python
def get_predictions(model, x):
    predictions=model.predict(x)
    pred=[]
    for prediction in predictions:
        pred.append(np.argmax(prediction))
    return pred
```

```python
def get_data(data):
    for number in os.listdir('Handwritten Numbers'):
        path=os.path.join('Handwritten Numbers', number)
        image=cv.imread(path, cv.IMREAD_GRAYSCALE)
        data.append(image)
    return data
data=[]
get_data(data)
print(data)
```

```python
X=np.array(data).reshape(-1,28,28,1)
X=X/255
```

```python
def evaluate_predictions(predictions):
    for prediction,img in zip(predictions,data):
        print_image(img)
        print("Model Predictions", prediction)
```

```python
pred=get_predictions(cnn_model, X)
```

```python
evaluate_predictions(pred)
```
